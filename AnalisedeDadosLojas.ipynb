#1. Análise do faturamento


```python
faturamento_por_loja = df_total.groupby('Loja')['Faturamento_Item'].sum().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='Loja', y='Faturamento_Item', data=faturamento_por_loja, palette='viridis')
plt.title('Faturamento Total por Loja')
plt.xlabel('Loja')
plt.ylabel('Faturamento Total (R$)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# 2. Vendas por Categoria


```python
vendas_por_categoria_total = df_total['Categoria do Produto'].value_counts()

plt.figure(figsize=(10, 8))

plt.pie(vendas_por_categoria_total, labels=vendas_por_categoria_total.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Distribuição Percentual de Vendas por Categoria de Produto (Todas as Lojas)')
plt.axis('equal')
plt.show()
```

# 3. Média de Avaliação das Lojas

```python
media_avaliacao_por_loja = df_total.groupby('Loja')['Avaliação da compra'].mean().reset_index()

plt.figure(figsize=(10, 6))

sns.barplot(x='Avaliação da compra', y='Loja', data=media_avaliacao_por_loja, palette='magma', orient='h')
plt.title('Média de Avaliação das Lojas')
plt.xlabel('Média de Avaliação (1-5)')
plt.ylabel('Loja')
plt.xlim(0, 5)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python
produtos_mais_vendidos = df_total['Produto'].value_counts().head(5)

plt.figure(figsize=(10, 6))
sns.barplot(x=produtos_mais_vendidos.index, y=produtos_mais_vendidos.values, palette='Greens_r')
plt.title('Top 5 Produtos Mais Vendidos (Todas as Lojas)')
plt.xlabel('Produto')
plt.ylabel('Quantidade Vendida')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

produtos_menos_vendidos = df_total['Produto'].value_counts().tail(5)

plt.figure(figsize=(10, 6))
sns.barplot(x=produtos_menos_vendidos.index, y=produtos_menos_vendidos.values, palette='Reds_r')
plt.title('Top 5 Produtos Menos Vendidos (Todas as Lojas)')
plt.xlabel('Produto')
plt.ylabel('Quantidade Vendida')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# 5. Frete Médio por Loja

## Relatório de Análise de Desempenho das Lojas

### Objetivo

O objetivo principal desta análise foi identificar a loja com o **menor desempenho** entre as quatro opções fornecidas, com o propósito de recomendá-la para venda, liberando capital e foco para um novo empreendimento.

### Metodologia

Analisamos os seguintes critérios para cada loja:

* **Faturamento Total**: A soma do preço dos produtos e do frete.
* **Frete Médio**: O custo médio do frete por item.
* **Avaliação Média**: A média das avaliações dos clientes (em uma escala de 1 a 5).
* **Vendas por Categoria**: A distribuição e o foco das vendas em diferentes categorias de produtos.
* **Produtos Vendidos**: Uma visão geral dos produtos que mais e menos contribuíram para as vendas.

---



### Análise e Justificativa da Recomendação

Após a análise aprofundada dos dados, a **Loja 1** se destaca como a candidata ideal para venda devido ao seu desempenho inferior em múltiplos aspectos críticos:

1.  **Faturamento Total Mais Baixo**: A Loja 1 apresentou o **menor faturamento total** (R$ 2.335,95) entre todas as lojas. Isso indica um volume de vendas e/ou valor de produtos significativamente menor, o que impacta diretamente a sustentabilidade e o potencial de crescimento do negócio.

2.  **Pior Avaliação Média dos Clientes**: Com uma média de apenas **3,0 estrelas**, a Loja 1 tem o **menor índice de satisfação do cliente**. Avaliações negativas podem refletir problemas na qualidade dos produtos, atrasos na entrega, falhas no atendimento ou uma combinação desses fatores. Manter um negócio com baixa satisfação do cliente é um desafio e exige investimentos consideráveis para reverter a percepção pública e a reputação.

3.  **Falta de Destaque em Produtos ou Categorias**: Embora tenha diversidade de categorias, a Loja 1 não apresenta nenhum produto ou categoria que se destaque em volume de vendas. Isso pode sugerir que os produtos oferecidos não estão ressoando bem com o mercado ou que a loja não conseguiu criar um nicho de sucesso.

Mesmo que o frete médio da Loja 1 não seja o mais alto, a combinação de baixo faturamento e, mais criticamente, a insatisfação do cliente, tornam-na a opção menos eficiente e com maior risco para manter ou investir. Investir em um novo empreendimento exige foco e recursos, e lidar com os problemas de uma loja com desempenho tão fraco desviaria energia valiosa.

---

### Recomendação Final

Com base na análise de vendas, desempenho e avaliações, a **Loja 1 é a loja recomendada para ser vendida**.

Ao desinvestir na Loja 1, o Senhor João poderá realocar capital e dedicação para o seu novo empreendimento, focando em um modelo de negócio que demonstre maior eficiência e potencial de sucesso, como observado em outras lojas (por exemplo, a **Loja 4**, que se destacou em avaliação e diversidade, ou a **Loja 3**, com o maior faturamento).
